```python
def rename_files(image_dir):
    import os
    for root, dirs, files in os.walk(image_dir):
        for file in files:
            if file.endswith(".chip.jpg"):
                os.rename(os.path.join(root, file), os.path.join(root, file.replace(".chip.jpg", "")))
                print("Renamed: ", os.path.join(root, file), " to ", os.path.join(root, file.replace(".chip.jpg", "")))
rename_files("/media/gnort/HDD6/Downloads_New/Compress/UTKface_aligned/UTKFace")
```

```python
import os
import random
image_dir = sorted(os.listdir("/media/gnort/HDD6/Downloads_New/Compress/UTKface_aligned/UTKFace"))
race_images = []
for image_name in image_dir:
    if image_name.split("_")[2] == "1" and len(image_name.split(" ")) == 1:
        race_images.append(image_name)
random.seed(31337)
# race_list = random.sample(race_images, 5500)
race_list = race_images
print(len(race_list))
```

```python
def create_txt_file(image_list, attr, file_name):
    if os.path.exists(file_name):
        os.remove(file_name)
    s = " 0"
    if attr == "age":
        pos = 0
    elif attr == "gender":
        pos = 1
    elif attr == "race":
        pos = 2
    else:
        raise ValueError("Invalid attribute")
    with open(file_name, "w") as f:
        for image in image_list:
            if (len(image.split(" "))) > 1:
                continue
            attr_value = image.split("_")[pos]
            if pos == 1:
                attr_value = str(int(attr_value) + 1)
                f.write(image + s + " " + attr_value + s * 4 + "\n")
            if pos == 2:
                attr_value = str(int(attr_value) + 1)
                f.write(image + " " + attr_value + s * 5 + "\n")
    print(len(image_list))
            
create_txt_file(race_list,
                "race",
                "/media/gnort/HDD6/Study/face-analysis-lightning/dataset/face/image_lists/added_race.txt")
```

```python
def copy_images(image_list, source_dir, destination_dir):
    for image in image_list:
        os.system("cp " + source_dir + image + " " + destination_dir)
        print("Copied " + image)
        
source_dir = "/media/gnort/HDD6/Downloads_New/Compress/UTKface_aligned/UTKFace/"
destination_dir = "/media/gnort/HDD6/Study/face-analysis-lightning/dataset/face/aligned_faces_added_race/"
image_list = race_list
copy_images(image_list, source_dir, destination_dir)
```

```python
def merge_two_txt_files(file1, file2, file3):
    if os.path.exists(file3):
        os.remove(file3)
    with open(file1, "r") as f1, open(file2, "r") as f2, open(file3, "w") as f3:
        for line in f1:
            f3.write(line)
        for line in f2:
            f3.write(line)
    print("Merge done!")
file1 = "/media/gnort/HDD6/Study/face-analysis-lightning/dataset/face/image_lists/face_train.txt"
file2 = "/media/gnort/HDD6/Study/face-analysis-lightning/dataset/face/image_lists/added_race.txt"
file3 = "/media/gnort/HDD6/Study/face-analysis-lightning/dataset/face/image_lists/face_train_added_race.txt"
merge_two_txt_files(file1, file2, file3)
```
