```python
df = pd.read_csv("/media/gnort/HDD6/Study/face-analysis-lightning/dataset/face/image_lists/face_train.txt", sep=" ", header=None)
```

```python
indices_to_keep_1 = np.random.choice(df[df[1] == 1].index, 3000, replace=False)
indices_to_keep_0 = np.random.choice(df[df[1] == 0].index, 3000, replace=False)
indices_to_keep_2 = df[df[1] == 2].index

# Combine the selected indices for both values
indices_to_keep = np.concatenate([indices_to_keep_1, indices_to_keep_0, indices_to_keep_2])

# Keep only the selected indices and reset the index
df_filtered = df.loc[indices_to_keep].reset_index(drop=True)
```
